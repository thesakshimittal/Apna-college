```python
import numpy as np
```

```python
original = cv2.imread("/content/mona_lisa.jpg")
encrypted_image = cv2.imread("/content/encrypted_image.webp")
decrypted_image = cv2.imread("/content/decrypted image.jpeg")
```

```python
def encrypt(image):
  key = np.random.randint(0,256, image.shape)
  encrypted_image = np.zeros_like(image)
  for i in range(image.shape[0]):
    for j in range(image.shape[1]):
      encrypted_image[i,j] = image[i,j]^key[i,j]
  return encrypted_image, key
```

```python
def decrypt(encrypted_image):
  decrypted_image = np.zeros_like(encrypted_image)
  for i in range(encrypted_image.shape[0]):
    for j in range(encrypted_image.shape[1]):
      decrypted_image[i,j] = encrypted_image[i,j]^key[i,j]
  return decrypted_image
```

```python
encrypted_image, key = encrypt(original)
```

```python
plt.imshow(encrypted_image)
```

```python
decrypted_image = decrypt(encrypted_image)
```

```python
plt.imshow(decrypted_image)
```

```python
cv2.imwrite("/content/encrypted_image.webp",encrypted_image)
```

```python
cv2.imwrite("/content/decrypted image.jpeg",decrypted_image)
```
